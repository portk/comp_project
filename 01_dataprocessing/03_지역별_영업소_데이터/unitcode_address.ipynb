```python
import pandas as pd

import numpy as np

import json
```

```python
# 1. 파일 읽기
#파일 출처 : https://kimhongsi.tistory.com/entry/%ED%96%89%EC%A0%95%EA%B5%AC%EC%97%AD-%EC%BD%94%EB%93%9C-%EC%A0%95%EB%A6%AC-%EC%8B%9C%EB%8F%84-%EB%8B%A8%EC%9C%84-%EC%8B%9C%EA%B5%B0%EA%B5%AC-%EB%8B%A8%EC%9C%84
file2_name='행정동_기준_시군구_단위.xlsx' #행정동 기준의 시도코드, 시군구

# 파일출처 : https://www.data.go.kr/data/15085270/fileData.do
file1_name='한국도로공사_지사별 영업소 현황_20230630.csv'

#20240131 일일히 react 마커보고 작업한 것
file3_name='행정동시군구_마커_그룹화.xlsx'

sigun=pd.read_excel(file2_name)
address=pd.read_csv(file1_name, encoding='euc-kr')
marker=pd.read_excel(file3_name, dtype={'unitCode':str})
```

```python
# 1-1. 데이터 전처리
# 데이터프레임marker 전처리하기 : 0이 사라지면 다시 채워주기
marker.unitCode = np.where(marker.unitCode.str.len() > 2,marker.unitCode, '0'+marker.unitCode)
```

```python
js1_pd=pd.read_json('1.json')['list']
js2_pd=pd.read_json('2.json')['list']
js3_pd=pd.read_json('3.json')['list']
js4_pd=pd.read_json('4.json')['list']
js5_pd=pd.read_json('5.json')['list']
js6_pd=pd.read_json('6.json')['list']

js_series=pd.concat([js1_pd,js2_pd,js3_pd,js4_pd,js5_pd,js6_pd])

js_series=js_series.reset_index(drop=True)

tmp=list(js_series)
# len(tmp)

js_pd=pd.DataFrame(tmp)

columns = ['unitName', 'unitCode', 'routeNo', 'routeName', 'xValue', 'yValue']
js_pd=js_pd[columns]

# js_pd.isnull().sum()

js_pd['unitCode']=js_pd['unitCode'].str.strip()
js_pd
```
